# Projeto: Criar um script em Python que leia o Wikipedia, descubra as capitais brasileiras, e delas busque novamente no Wikipedia e entregar um resumo. Exibir em um DataFrame.


### Para iniciar, como o intuito do projeto é acessar uma página web, escolhi a biblioteca requests, para fazer solicitações HTTP, permitindo que eu envie requisições GET, POST e outras para interagir com recursos da web. Em conjunto usei a BeautifulSoup para análise de páginas da web e extração de informações de documentos HTML e XML. Estou utilizando o google colab, por esse motivo não é necessária a instalação manual das bibliotecas , apenas importar. Ao longo do projeto, adicionarei comentários breves para explicar o passos.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://pt.wikipedia.org/wiki/Lista_de_capitais_do_Brasil_por_%C3%A1rea'
```

*Leitura da url*

```python
requisicao = requests.get(url)
```

Verificação do código HTTP.(
 200 -> Ok )

```python
requisicao.status_code
```

Armazenei o conteúdo obtido pela url em uma variavel

```python
conteudo_html = requisicao.content
```

```python
type(conteudo_html)
```

Precisei transformar o tipo da variavel para que seja um objeto da classe BeautifulSoup e assim podermos usar os metodos da lib

```python
wikipedia = BeautifulSoup(conteudo_html, 'html.parser')
```

```python
type(wikipedia)
```

Verificar como está a variável, se podemos analisar as tags facilmente

```python
print(wikipedia)
```

Ao analisar o inspecionar e a variavel wikipedia , percebi que o que preciso é a tabela, portanto pegar somente a tag table, especificando a classe (se nao colocar o atributo class: pegaria a primeira tag table)

```python
wikipedia.find('table', attrs={'class': 'wikitable'})
```

Armazenei a tag table que eu queria em uma variável

```python
tabela_capitais = wikipedia.find('table', attrs={'class': 'wikitable'})
```

Visualização da nossa tag table armazenada na variavel tabela capitais.

```python
print(tabela_capitais)
```

Por mais que consigamos ler e analisar as tags para trabalhar, tem um método muito interessante que facilita ainda mais a leitura da variável, usando identação e deixando as separações mais claras

```python
print(tabela_capitais.prettify())
```

Hora de procurar o nome das capitais e os respectivos HREFs. Para posterior consulta na wikipedia. Para isso criei duas listas para armazenar os valores durante o loop

```python
capitais = []
links = []

for cada_tr in tabela_capitais.findAll('tr')[1:]: #buscar todas as tags 'tr' a partir da Segunda(indice 1) para pular o cabeçalho contido no site

  a_tag = cada_tr.findAll('td')[1].find('a') # de todas as tags 'td' a partir do segundo td, encontrar a primeira tag 'a' que corresponde ao nome da capital

  if a_tag: # encontrando a tag 'a' executar
    nome_capital = a_tag.text # obter o texto da tag e armazenar em uma variável
    link_href = a_tag.get('href') # obter o HREF da tag e armazenar em uma variável
    if nome_capital and nome_capital.lower() != 'none': # verificar se não está com tipo None
                capitais.append(nome_capital) # armazena o nome obtido , na lista capitais
                links.append(link_href) # armazena o HREF obtido , na lista links

# agora tenho uma lista do nome das capitais e outra com os links das respectivas capitais
print(capitais)
print(links)
```

Verificando a quantidade de capitais pelo metodo len. Diz o tamanho da lista. Justamente a quantidade total de capitais. Que são os 26 estados e o Distrito Federal.

Vou precisar acessar dois valores por vez em loop de pesquisa do site, portanto é interessante zipar as duas listas em um dicionário para cada item eu conseguir o acesso a dois valores(chave e valor) e nao precisar de dois for(um para cada lista).

```python
capital_com_link = dict(zip(capitais, links))
print(capital_com_link)
```

Ideia do for: Para cada item(Capital e respectivo link) acesso o link , procuro o segundo parágrafo(observado pelo inspecionar do site que seria o parágrafo do resumo) obtenho o texto do parágrafo e armazeno na minha lista de resumos declarada antes do loop.

```python
resumo = []

for capital, link in capital_com_link.items(): # percorrendo cada item(chave + valor) do dicionario

    url2 = f'https://pt.wikipedia.org{link}' # visita o site que contém informações sobre a respectiva capital

    requisicao = requests.get(url2) # armazena a requisicao do site em uma variável

    if requisicao.status_code == 200: # verifica o código HTTP gerado. (Lembrando -> 200 = Ok).

        pagina = BeautifulSoup(requisicao.text, 'html.parser') # usei o beautifulSoup para pegar o texto da requisicao, analisar o HTML com o analisador "parser" (Acessa o HTML de forma inteligente)

        paragrafos = pagina.find_all('p') # acessa todas as tags 'p' contida no site.

        if len(paragrafos) >= 2: # analisando o  inspecionar assim como na variável wikipedia, consegui perceber que a segunda tag 'p' correspondia ao breve resumo da capital. por isso o if(evitar paginas com menos de 2 tags 'p')

            texto_paragrafo = paragrafos[1].get_text() # obter o texto da segunda tag 'p'


            resumo.append(texto_paragrafo) # incluir o texto na lista de resumos

            # print(f'{capital}\nResumo de {capital}:\n {texto_paragrafo}\n') print para teste e verificar se ocorreu tudo certo na aquisição do texto
        else:
                print(f'Não foi possível encontrar o segundo parágrafo para {capital}.\n') # caso a página tivesse menos de 2 paragrafos

    else:
        print(f"Erro, acesso à página não obtido. Código HTTP: {requisicao.status_code}") # mensagem similar ao tratamento de erro, para exibir o código HTTP obtido, caso seja diferente de 200(OK)
```

Mesma ideia do loop de aquisição da capital. Porém os estados eu só quero pegar o texto, para organizar no DataFrame.

```python
estados = []

for cada_tr in tabela_capitais.findAll('tr')[1:]: # passando indice 1 adiante para pular o cabeçalho
                # acessar o quarto td e entao acessar a primeira tag a
  a_tag = cada_tr.findAll('td')[3].find('a', recursive=False)  # recursive false, apenas as tags 'a' que sejam filhas diretas do td(não tenham outra tag pai). Pois tem uma tag no estado Pará que está dentro de uma span

  if a_tag:
    nome_estado = a_tag.text # se tiver a tag 'a' , obter o texto e armazenar em uma variável
    if nome_estado and nome_estado.lower() != 'none': # verificar se a variável é compatível com o tipo
                estados.append(nome_estado) # incluir na lista

print(estados)
```

Verificação da quantidade de elementos da lista

Criei uma variável para armazenar o dataframe gerado pelas listas capitais, estados e resumo. Porém usando um dicionário. Com cada chave sendo o nome da cooluna, e seus valores sendo os valores da tabela

```python
df = pd.DataFrame({'Nome da Capital': capitais, 'Estado': estados , 'Resumo': resumo})
```
